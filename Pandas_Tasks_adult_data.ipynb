summary of the DataFrame

```python
import pandas as pd
df = pd.read_csv('adult.csv')
df.info()
```

```python
display(df.describe())
```

Display the column names of the DataFrame.

```python

display(df.columns)
```

unique values in the 'workclass' column.

```python

display(df['workclass'].unique())
```

*Create* a new column 'income_binary' where <=50K is 0 and >50K is 1.

```python
df['income_binary'] = df['income'].apply(lambda x: 0 if x == '<=50K' else 1)
display(df.head())
```

Create a new column 'age_group' categorizing ages into 'Youth' (< 30), 'Adult' (30-60), and 'Senior' (60+).

```python

df['age_group'] = pd.cut(df['age'], bins=[0, 30, 60, float('inf')], labels=['Youth', 'Adult', 'Senior'])
display(df.head())
```

Rename the 'educational-num' column to 'education_num'.

```python

df = df.rename(columns={'educational-num': 'education_num'})
display(df.head())
```

Replace the missing values represented by '?' in 'workclass' and 'occupation' columns with 'Unknown'.

```python

df['workclass'] = df['workclass'].replace('?', 'Unknown')
df['occupation'] = df['occupation'].replace('?', 'Unknown')
display(df.head())
```

Calculate the average age for each workclass.

```python

display(df.groupby('workclass')['age'].mean())
```

Find the total capital-gain for each gender.

```python

display(df.groupby('gender')['capital-gain'].sum())
```

Create 5 different charts and write your observations from the chart as comment

```python

# Chart 1: Distribution of Age
import matplotlib.pyplot as plt
plt.figure(figsize=(8,5))
plt.hist(df['age'], bins=30, edgecolor='black', alpha=0.7)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
# Observation: The age distribution is right-skewed. Most people are between 2050 years old.

# Chart 2: Income proportion by Gender
plt.figure(figsize=(6,5))
df.groupby('gender')['income_binary'].mean().plot(kind='bar', color=['lightblue', 'salmon'])
plt.title('Proportion Earning >50K by Gender')
plt.ylabel('Proportion')
plt.xticks(rotation=0)
plt.ylim(0, 0.4)
for i, v in enumerate(df.groupby('gender')['income_binary'].mean()):
    plt.text(i, v + 0.01, f'{v:.3f}', ha='center')
plt.show()
# Observation: Males have ~3 times higher chance of earning >50K than females.

# Chart 3: Hours per week by Education (boxplot)
plt.figure(figsize=(12,6))
df.boxplot(column='hours-per-week', by='education', rot=45)
plt.title('Hours per Week by Education Level')
plt.suptitle('')
plt.xlabel('Education')
plt.ylabel('Hours per Week')
plt.show()
# Observation: People with Prof-school and Doctorate tend to work more hours (median ~4550).

# Chart 4: Marital Status distribution
plt.figure(figsize=(8,8))
df['marital-status'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Marital Status Distribution')
plt.ylabel('')
plt.show()
# Observation: Nearly half the dataset is "Married-civ-spouse", followed by "Never-married".

# Chart 5: Capital Gain vs Age (scatter)
plt.figure(figsize=(10,6))
plt.scatter(df['age'], df['capital-gain'], alpha=0.5, c=df['income_binary'], cmap='coolwarm')
plt.colorbar(label='Income >50K (1) / <=50K (0)')
plt.title('Capital Gain vs Age')
plt.xlabel('Age')
plt.ylabel('Capital Gain')
plt.show()
# Observation: Almost all capital-gain = 0. High capital gains (>20k) occur mostly in age 3070 and strongly correlate with income >50K.
```
